## Moving 43 Random Images from train to test Folder

```python
from pathlib import Path
import random, shutil

# Create test directories if they don't exist
Path("License-Plate-Data/test/images").mkdir(parents=True, exist_ok=True)
Path("License-Plate-Data/test/labels").mkdir(parents=True, exist_ok=True)

# Get base names without extension
files = []
for p in Path("License-Plate-Data/train/images").glob("*.png"):
    files.append(p.stem)

# Select 43 random
random.seed(42)
selected = random.choices(files, k=43)

# Move files
for name in selected:
    shutil.move(f"License-Plate-Data/train/images/{name}.png", f"License-Plate-Data/test/images/{name}.png")
    shutil.move(f"License-Plate-Data/train/labels/{name}.txt", f"License-Plate-Data/test/labels/{name}.txt")
```

## Verifying Dataset Split Counts

```python
from pathlib import Path

for folder in ['train', 'valid', 'test']:
    imgs = len(list(Path(f"License-Plate-Data/{folder}/images").glob("*")))
    lbls = len(list(Path(f"License-Plate-Data/{folder}/labels").glob("*")))
    print(f"{folder}: {imgs} images, {lbls} labels")
```
