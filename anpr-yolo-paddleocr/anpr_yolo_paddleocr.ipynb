## License Plate Detection and OCR Function

```python
def read_plate(image_path, model="best.pt"):
    """
    Detect and read license plates from an image using YOLO and PaddleOCR.

    Args:
        image_path (str): Path to the input image
        model (str): Path to YOLO model weights (default: "best.pt")

    Returns:
        list: List of tuples containing (plate_number, detection_confidence, ocr_confidence)
              Returns empty list if no plates detected
    """
    # Load YOLO model and run detection
    model = YOLO(model)
    results = model.predict(image_path, verbose=False)

    # Read the image for cropping
    img = cv2.imread(image_path)

    plates = []

    # Iterate through each detected bounding box
    for box in results[0].boxes:
        # Extract bounding box coordinates
        x1, y1, x2, y2 = box[0].xyxy[0].tolist()
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        # Get detection confidence score
        box_confidence = box[0].conf.item()

        # Crop the detected plate region
        cropped_plate = img[y1:y2, x1:x2]

        # Initialize PaddleOCR for text extraction
        ocr = PaddleOCR(lang="en", device="gpu", use_textline_orientation=True)
        ocr_result = ocr.predict(cropped_plate)

        # Extract text and OCR confidence if text is detected
        if ocr_result[0]['rec_texts']:
            plate_number = ocr_result[0]['rec_texts'][0]
            ocr_confidence = ocr_result[0]["rec_scores"][0]
            plates.append((plate_number, box_confidence, ocr_confidence))

    return plates
```

## Display Results from `read_plate()` Function

```python
# Read and display all license plated from the image
all_plates = read_plate("/content/7962.jpg")

# Print the number of plates found
print(f"Found {len(all_plates)} plates:")

# Display each plate with detection and OCR confidence scores
for plate, det_conf, ocr_conf in all_plates:
    print(f"  {plate} (detection: {det_conf:.2%}, OCR: {ocr_conf:.2%})")

# Visualize the detected plates with bounding boxes
results = model.predict("/content/7962.jpg")
plt.imshow(results[0].plot()[..., ::-1])
plt.axis("off")
```
