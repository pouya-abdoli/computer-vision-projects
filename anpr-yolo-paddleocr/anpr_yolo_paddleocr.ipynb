## Initialize PaddleOCR and Run on Sample Image

```python
# Suppress verbose logs from PaddleX, show only warnings and errors
import logging
logging.getLogger("paddlex").setLevel(logging.WARNING)
import matplotlib.pyplot as plt

from paddleocr import PaddleOCR

ocr = PaddleOCR(lang="en",
                device="gpu",
                use_textline_orientation=True) # Auto-detects and corrects rotated text lines (0° or 180°) for better accuracy.

img_path = r"/content/2015-BMW-320d-xDrive-Touring-test-drive-67.jpg"

result = ocr.predict(img_path)
```

```python
plt.imshow(plt.imread(img_path))
plt.axis("off")
plt.show()

# Get text and confidence scores together
for text, score in zip(result[0]['rec_texts'], result[0]['rec_scores']):
    print(f"{text} (confidence: {score:.2%})")
```

## Full Pipeline: YOLO Plate Detection + PaddleOCR Text Extraction

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
from paddleocr import PaddleOCR

img_path = r"/content/images.jpg"

# Load the pre-trained YOLO model (best.pt)
model = YOLO("best.pt")

# Run inference on the input image to detect plates
results = model.predict(img_path)

# Extract bounding box coordinates and confidence score for the first detected plate
boxes = results[0].boxes
if len(boxes) > 0:
    x1, y1, x2, y2 = boxes[0].xyxy[0].tolist()
    confidence = boxes[0].conf.item()

    # Convert coordinates to intergers for image slicing
    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

    # Load image with OpenCV and crop the detected plate region
    img = cv2.imread(img_path)
    cropped_plate = img[y1:y2, x1:x2]

    # Visulaize: Show original detection + cropped late sidy-by-side
    plt.figure(figsize=(17, 15))

    plt.subplot(121)
    plt.imshow(results[0].plot()[..., ::-1]) # Convert BGR to RGB for display
    plt.axis("off")
    plt.title("Detected Plate")

    plt.subplot(122)
    plt.imshow(cropped_plate[..., ::-1]) # Convert BGR to RGB for display
    plt.axis("off")
    plt.title(f'Cropped Plate (conf: {confidence:.2%})')

    plt.tight_layout()
    plt.show()

    # Initialize PaddleOCR for text extraction
    ocr = PaddleOCR(lang="en",
                    device="gpu",
                    use_textline_orientation=True) # Auto-detects and corrects rotated text lines (0° or 180°) for better accuracy.

    # Run OCR on the cropped plate image
    ocr_result = ocr.predict(cropped_plate)

    # Print extracted text and confidence scores
    for text, score in zip(ocr_result[0]['rec_texts'], ocr_result[0]['rec_scores']):
        print(f"Plate number: {text} (confidence: {score:.2%})")
```
