```python
import cv2

# Initialize webcame (0 = defualt camera)
cap = cv2.VideoCapture(0)

# Loop continuously to process each vide frame
while True:
    
    # Read frame from webcame (ret = success boolean, frame = image)
    ret, frame = cap.read()

    # Flip frame horizontally
    frame = cv2.flip(frame, 1)

    # Convert frame to grayscale (Haar Cascade requires)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Load Haar Cascade XML file for frontal face detection
    haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

    # Detect faces
    faces_rec = haar_cascade.detectMultiScale(image=gray,
                                              scaleFactor=1.1,
                                              minNeighbors=15)

    # Draw rectangle around each deteced face
    for (x, y, w, h) in faces_rec:
        cv2.rectangle(img=frame, pt1=(x, y), pt2=(x + w, y + h), color=(0, 255, 0), thickness=2)

    # Display number of faces detented on the frame
    cv2.putText(img=frame,
                text=f"Faces detected: {len(faces_rec)}",
                org=(170, 50),                              # Position of bottom-left corner of text
                fontFace=cv2.FONT_HERSHEY_DUPLEX,           # Fony style
                fontScale=1,                                # Text size
                color=(0, 0, 255),
                thickness=1)
    
    # Show the frame 
    cv2.imshow("Face Detection", frame)
    
    # Break loop if "e" key is pressed
    # # & 0xFF extracts last 8 bits for cross-platform compatibility (prevents weird OS-specific key codes)
    if cv2.waitKey(1) & 0xFF == ord("e"):
        break

# Release webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```

<summary><b>`& 0xFF` Explained from Scratch</b> (click to expand)</summary>

### 1. What is `&`?
Bitwise AND operator - compares two numbers bit by bit.
Rule: Both bits must be 1 -> result is 1. Otherwise -> 0.

Example: `5 & 3`
5 in binary:  101
3 in binary:  011
Result:       001 (which is 1)

### 2. What is `0xFF`?
- `0x` means hexadecimal
- `FF` in hex = 255 in decimal = `11111111` in binary (8 ones)

### 3. What does `& 0xFF` do?
Keeps only the last 8 bits (1 byte), zeros out everything else.

Number: 1048608 (what macOS returns for 'e')

Binary: `100000000000000100000` (20 bits)

`0xFF`: `00000000000000011111111` (last 8 bits = 1)

After `&` operation:
Result: Last 8 bits only -> `01100001` = 101 decimal = 'e' ✅

### 4. Why use it?
Without `& 0xFF`: `1048608 == 101` -> False ❌

With `& 0xFF`: `101 == 101` -> True ✅

### Summary
> `& 0xFF` strips away extra bits, keeping only the last 8 bits (0-255),
> ensuring you get the correct key code on any operating system.
